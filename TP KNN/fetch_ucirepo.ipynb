```python
x_data = data.data.features
y_data = data.data.targets
```

```python
x = x_data.values
```

```python
y = y_data.values
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=5)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)   
x_test = sc.transform(x_test) 
```

```python
def distance(x1,x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
def kni(x_train, xi, k):
    distances = []
    for i in range(len(x_train)):
        d = distance(xi,x_train[i])
        distances.append([d,i])
    distances.sort()
    distances = np.array(distances [:k])
    return distances[:,-1]
```

```python
def predict(x_train, y_train, X, k):
    pred = []
    for xi in X:
        knn_index = kni(x_train, xi, k)
        freq = {}
        for i in knn_index:
            i = int(i)
            label = y_train[i][0]
            if label in freq:
                freq[label] += 1
            else:
                freq[label] = 1
        # choisir la classe la plus fréquente
        f = list(freq.keys())[0]
        for i in freq:
            if freq[i] > freq[f]:
                f = i
        pred.append(f)
    return pred
```

```python
y_pred = predict(x_train, y_train, x_test, k = 10)
```

```python
accuracy_score(y_test, y_pred)
```
